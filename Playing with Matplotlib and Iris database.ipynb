```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris['DESCR'])
```

```python
x = iris.data.T[0]
y = iris.data.T[1]
s = 100*iris.data.T[3]
c = iris.target

fig, ax = plt.subplots(figsize = (14,6))
scatter = ax.scatter(x, y, s=s, c=c, cmap= 'Set1', alpha = 0.6)
plt.xlim(4.5, 8.5)
plt.title('Sepal length, width and petal length  per class')
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')


legend1 = ax.legend(*scatter.legend_elements(), loc="upper right", title="Classes")
ax.add_artist(legend1)
plt.show()
```
